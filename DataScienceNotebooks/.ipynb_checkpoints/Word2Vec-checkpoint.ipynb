```python
import pandas.io.sql as psql
import re
import psycopg2
from gensim.models.word2vec import Word2Vec
```

```python
import json
with open('/Users/krishna/MOOC/smart-city/login.json') as data_file:
    db = json.load(data_file)

conn = psycopg2.connect(database=db['dbname'], user=db['user'], password=db['password'], host=db['host'], port=db['port'])
```

```python
dataframe = psql.frame_query("select text from twitter.tweets limit 50000",conn)
```

```python
dataframe.head()
```

```python
def processTweet(tweet):
    #Convert to lower case
    tweet = tweet.lower()
    #Convert www.* or https?://* to URL
    tweet = re.sub('((www\.[^\s]+)|(https?://[^\s]+))','URL',tweet)
    #Convert @username to AT_USER
    tweet = re.sub('@[^\s]+',' ',tweet)
    #Remove additional white spaces
    tweet = re.sub('[\s]+', ' ', tweet)
    #Replace #word with word
    tweet = re.sub(r'#([^\s]+)', r'\1', tweet)
    #trim
    tweet = tweet.strip('\'"')
    return tweet.split(" ")
x_train = processTweet(dataframe['text'])
x_train.head()
```

```python
min_count = 2
size = 50
window = 4
model = Word2Vec(dataframe['text'], min_count=min_count, size=size, window=window)
```

```python
w2v = Word2Vec(size=300, min_count=30,window = 4)
w2v.build_vocab(x_train)
```
